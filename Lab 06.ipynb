```python
def count_vowel(a):
    count = 0
    for i in a:
        if i == 'a' or i == 'e'or i == 'i' or i == 'o' or i == 'u':
            count += 1
    if count == 0:
        print("No Vowels are Present in the input")
    else:
        print("Number of vowels present in the input:",count)

def vowel_individualcount(a):
    counta=0
    for i in a:
        if i=='a':
            counta+=1
    print("-->a")
    print("number of 'a' present in the input:",counta)
    print("Percentage of 'a' present in the input:",counta/len(a)*100,"%")
    counte=0
    for i in a:
        if i=='e':
            counte+=1
    print("-->e")
    print("number of 'e' present in the input:",counte)
    print("Percentage of 'e' present in the input:",counte/len(a)*100,"%")
    counti=0
    for i in a:
        if i=='i':
            counti+=1
    print("-->i")
    print("number of 'i' present in the input:",counti)
    print("Percentage of 'i' present in the input:",counti/len(a)*100,"%")
    counto=0
    for i in a:
        if i=='o':
            counto+=1
    print("-->o")
    print("number of 'o' present in the input:",counto)
    print("Percentage of 'o' present in the input:",counto/len(a)*100,"%")
    countu=0
    for i in a:
        if i=='u':
            countu+=1
    print("-->u")
    print("number of 'u' present in the input:",countu)
    print("Percentage of 'u' present in the input:",countu/len(a)*100,"%")

def vowel_percentagecheck(a):
    count = 0
    for i in a:
        if i == 'a' or i == 'e'or i == 'i' or i == 'o' or i == 'u':
            count += 1
    d=count/len(a)*100
    print("Percentage of vowels present in the input:",d,"%")

a=input("Enter yout input")
print("*"*30)
print("The given input:",a)
print("*"*30)
count_vowel(a)
print("*"*30)
print("Below are the vowels present in the input:")
vowel_individualcount(a)
print("*"*30)
vowel_percentagecheck(a)
print("*"*30)
```
