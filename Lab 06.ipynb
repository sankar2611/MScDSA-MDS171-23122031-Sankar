```python
#user defined functions:
def count_totalvowesl(a): #user defined function for total number of vowels present.
   count = 0
   for i in a:
    if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
        count+=1
   if count == 0:
      print('No vowels found')
   else:
       print("Total vowels present in the input are :", count )

def vowel_individualcount(a): # function defined for calculating individual vowel"s presence in the oinput and its count.
   counta = 0
   for i in a:
      if i == 'a':
         counta +=1 
   print("=> a")
   print("total count of 'a' :",counta)
   s=(counta/len(a))*100
   print("percentage of 'a' present in the input:",s,"%")
   counte = 0
   for y in a:
      if y == 'e':
         counte +=1 
   print("=> e")
   print("total count of 'e' :",counte)
   s=(counte/len(a))*100
   print("percentage of 'e' present in the input:",s,"%")
   counti = 0
   for z in a:
      if z == 'i':
         counti +=1 
   print("=> i")
   print("total count of 'i' :",counti)
   s=(counti/len(a))*100
   print("percentage of 'i' present in the input:",s,"%")
   counto = 0
   for w in a:
      if w == 'o':
         counto +=1 
   print("=> o")
   print("total count of 'o' :",counto)
   s=(counto/len(a))*100
   print("percentage of 'o' present in the input:",s,"%")
   countu = 0
   for t in a:
      if t == 'u':
         countu +=1 
   print("=> u")
   print("total count of 'u' :",countu) 
   s=(countu/len(a))*100
   print("percentage of 'u' present in the input:",s,"%")   

def percentage_calculateofvowels(a): #function for calculating the percentage of vowels present in the input
   count = 0
   for i in a:
    if i == 'a' or i == 'e' or i == 'i' or i == 'o' or i == 'u':
        count+=1
   d=count/len(a)*100
   print("Percentage of vowels present in the input is:",d,"%")
#main function:
a = input('Enter the string :')
print("The Entered input is:",a)
count_totalvowesl(a)
print("Below are the vowels present:")
vowel_individualcount(a)
percentage_calculateofvowels(a)
```
