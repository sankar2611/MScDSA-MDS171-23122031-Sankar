```python
def matrix_input():
    rows = int(input("Enter the number of rows: "))
    columns = int(input("Enter the number of columns: "))

    matrix = []

    print("Enter the entries row-wise:")

    for i in range(rows):
        row = []
        for j in range(columns):
            entry = int(input(f"Enter element at position ({i+1}, {j+1}): "))
            row.append(entry)
        matrix.append(row)

    return matrix

def matrix_multiplication(matrix1, matrix2):
    if len(matrix1[0]) != len(matrix2):
        print("Matrix multiplication is not possible. The number of columns in the first matrix must be equal to the number of rows in the second matrix.")
        return None

    result = []

    for i in range(len(matrix1)):
        row = []
        for j in range(len(matrix2[0])):
            entry = 0
            for k in range(len(matrix2)):
                entry += matrix1[i][k] * matrix2[k][j]
            row.append(entry)
        result.append(row)

    return result

def print_matrix(matrix):
    for row in matrix:
        for element in row:
            print(element, end=" ")
        print()

def main():
    print("Matrix 1:")
    matrix1 = matrix_input()

    print("Matrix 2:")
    matrix2 = matrix_input()

    result = matrix_multiplication(matrix1, matrix2)

    if result:
        print("\nMatrix 1:")
        print_matrix(matrix1)
        print("\nMatrix 2:")
        print_matrix(matrix2)
        print("\nResult of matrix multiplication:")
        print_matrix(result)

if __name__ == "_main_":
    main()
```
