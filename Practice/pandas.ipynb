create a class pandas_name that will help you to read a csv file and print the top/bottom n rows in the file........
class pandas_name:
                init()
                        stores the data
                read csv()
                        reads a csv file 
                head()
                        prints the top n rows 
                tail()
                        prints the bottom n rows

```python
class pandas_sankar():
    def __init__(self):
        self.filename = ""
        self.filecontent = ""

    def read_csv(self, filename):
        self.filename = filename
        with open(filename, "r+") as file:  
            self.filecontent = file.readlines()

    def head(self, num_of_rows):
        for i in range(0, num_of_rows):
            print(self.filecontent[i]) 

    def tail(self, num_of_rows):
        for i in range(len(self.filecontent) - num_of_rows, len(self.filecontent)):  
            print(self.filecontent[i])  

pd = pandas_sankar()
pd.read_csv("dropout-ratio-2012-2015.csv")
pd.head(4)
pd.tail(3)
```

```python
import os
class pandas_sankar:
    def __init__(self):
        self.data_Dict={"dict":[]}
    
    def read_CSV(self,fileName):
        if os.path.isfile(fileName):
            for i in open(fileName,"r+").readlines():
                line=i.split(",")
                if line[2]=="":
                    pass
                else:
                    sub_Dict={"age":line[0],"expense":line[1],"familysize":line[2].strip()}
                    self.data_Dict["dict"].append(sub_Dict)
        else:
            print("File does not exist")
            return False

def describe(self,col):
        print(col)
        num=[]
        count=0
        sum=0
        mode=0
        flag1=0
        flag2=0
        for i in self.data_Dict["dict"]:
            if count!=0:
                sum+=int(i[col])
                num.append(int(i[col]))
            count+=1
        print("Count:",count-1)
        print("Mean:",sum/(count-1))
        for i in num:
            flag1=0
            for j in num:
                if i==j:
                    flag1+=1
            if flag1>flag2:
                flag2=flag1
                mode=i
        print("Mode:",mode)
        median=int(len(num)/2)-2
        print("Median:",num[median])

obj=pandas_sankar()
obj.read_CSV("PANDAS_EXCEL.csv")
#obj.head(3)
#print("\nTail")
#obj.tail(2)
column=int(input("Enter the column name needs to analysed - 1.Age 2.Expense 3.Family Size"))
columnName=["MATHS"]
obj.describe(columnName[column-1])
```
